```python
import pandas as pd

movies = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv')
credits = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv')
```

```python
#crear una base de dato sql

import sqlite3

#crear una coneccion 

conn = sqlite3.connect('../data/raw/movies.db')
```

```python
#llevar df a una tabla sql 
movies.to_sql('tabla_movies', conn, if_exists='replace', index=False)
credits.to_sql('tabla_creditos', conn, if_exists='replace', index=False)
```

```python
print(movies.columns)
print(credits.columns)
```

```python
#unir ambas tablas y crear una query

query = """
SELECT tc.movie_id, tm.title, tm.overview, tm.genres, tm.keywords, tc.cast, tc.crew
FROM tabla_ movies AS tm 
INNER JOIN tabla_creditos AS tc
ON tm.title = tc.title;
"""

df = pd.read_sql_query(query, conn)
conn.close()
df.head()
```
